```python
# Maquinas de Vectores de Soporte o SVM's

from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
```

```python
X = data.data
y = data.target
```

```python
import pandas as pd

df = pd.DataFrame(X, columns=data.feature_names)

df.head()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.5)
```

```python
# Se comporta mejor, con conjuntos de datosm de "pocos" registros
# pero con una gran cantidad de caracteristicas
from sklearn import svm
from sklearn.metrics import accuracy_score

# hiper-parametros: kernel y Gamma

modelo = svm.SVC() #Support Vector Classifier

modelo.fit(X_train, y_train)
predicciones = modelo.predict(X_test)
print("Resultado de exactitud: ", accuracy_score(y_test, predicciones))
```

```python
kernels = ['linear', 'rbf', 'sigmoid']

# gamma no afecta al kernel "linear"
gammas = [1, 0.1, 0.001, 0.0001, 0.00001]

for kernel in kernels:
  for gamma in gammas:
    modelo = svm.SVC(kernel=kernel, gamma = gamma) #Support Vector Classifier
    modelo.fit(X_train, y_train)
    predicciones = modelo.predict(X_test)
    print(f"{kernel} - {gamma} = Resultado de exactitud: ", accuracy_score(y_test, predicciones))
```

```python
modelo = svm.SVC(kernel='linear') #Support Vector Classifier
modelo.fit(X_train, y_train)
predicciones = modelo.predict(X_test)
accuracy_score(y_test, predicciones)
```

```python
from sklearn.metrics import confusion_matrix

pd.DataFrame(confusion_matrix(y_test, predicciones), columns=['Pred. No', 'Pred. Si'], index=['Real no','Real Si'])
```
